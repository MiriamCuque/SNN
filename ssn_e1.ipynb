```python

# 19th February 2025
# Homework 1, Miriam Cegarra Cuquerella

import numpy as np
import matplotlib.pyplot as plt

#Función para obtener la constante (k) de reacción en una reacción química de primer orden, y graficar el consumo de reactivo.
#Input: concentración inicial y final de reactivo (en M), tiempo transcurrido para el cambio de concentración (en s) y número de puntos para definir el intervalo de tiempo.
def reaccion_orden1(A0,At,t,puntos):

  k=(np.log(A0/At))/t                                                                      #Cálculo de la constante de velocidad a partir del tiempo y la relación entre concentraciones
  saltos=t/puntos                                                                          #Cálculo de intervalo de tiempos desde un inicio hasta a t según el número de puntos
  tiempos=np.arange(0,10*t,saltos)                                                         #Creación lista de tiempos
  concentraciones=[]                                                                       #Creación de lista de concentraciones
  for i in tiempos:                                                                        #Bucle para cálculo de concentraciones
    c=A0*np.exp(-k*i)                                                                      #Concentración final para una reacció de orden 1
    concentraciones.append(c)                                                              #Añadir a lista

  plt.plot(tiempos, concentraciones, label=f'Reacción 1° orden (k={k:.2} s⁻¹)', color='g') #Graficar concentración vs tiempo
  plt.xlabel("Tiempo (s)")                                                                 #Etiqueta eje X
  plt.ylabel("Concentración [A] (M)")                                                      #Etiqueta eje Y
  plt.title("Cinética de una reacción de primer orden")                                    #Título
  plt.legend()                                                                             #Leyenda
  plt.grid(True)                                                                           #Cuadrícula
  plt.show()

reaccion_orden1(1,0.5,10,10)
```
